```python
years = int(input("Enter the number of years: "))

days = years * 365
hours = days * 24
minutes = hours * 60
seconds = minutes * 60

print(f"{years} years is equal to:")
print(f"{days} days")
print(f"{hours} hours")
print(f"{minutes} minutes")
print(f"{seconds} seconds")
```

```python
years = int(input("Enter number of years: "))

leap_years = years // 4
normal_years = years - leap_years

days = normal_years * 365 + leap_years * 366
hours = days * 24
minutes = hours * 60
seconds = minutes * 60

print(f"Days: {days}")
print(f"Hours: {hours}")
print(f"Minutes: {minutes}")
print(f"Seconds: {seconds}")
```

First 100 units: ₹5/unit
Next 100 units: ₹7/unit
Above 200 units: ₹10/unit
(Add a fixed meter charge of ₹100)


```python

#First 100 units: ₹5/unit
#Next 100 units: ₹7/unit
#Above 200 units: ₹10/unit  (Add a fixed meter charge of ₹100)

units = int(input("Enter total units consumed: "))
bill = 100
if units <= 100:
    bill =bill + units * 5
elif units <= 200:
    bill += (100 * 5) + (units - 100) * 7
else:
    bill += (100 * 5) + (100 * 7) + (units - 200) * 10

print("Total Electricity Bill: ", bill)
```

wap to input a character and check whether it is a vowel or no.
match case [a,e,i,o,u,A,E,I,O,U]

```python
c = input("enter a character")
match c:
    case 'A'|'a':
        print("entered character is a vowel: A")
    case 'E'|'e':
        print("entered character is a vowel: E")
    case 'I'|'i':
        print("entered character is a vowel: I")
    case 'O'|'o':
        print("entered character is a vowel: O")
    case 'U'|'u':
        print("entered character is a vowel: U")
    case _:
        print("entered character is not a vowel")
```

```python
# count the number of vowels and consonants
c1=0
c2=0
char = input("Enter a string")
for i in char:
  match i:
    case 'a'|'e'|'i'|'o'|'u'|'A'|'E'|'I'|'O' |'U':
      c1=c1+1
    case _:
      c2=c2+1
print("Number of vowel characters is ",c1)
print("Number of consonents characters is ",c2)
```
